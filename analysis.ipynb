```python
import pandas as pd
import plotly.express as px
```

```python
pd.read_csv('results.csv')
```

```python
df = pd.read_csv('results_2.csv')
df.info()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df
```

```python
fig = px.scatter(df, x="timestamp", y="temperature", title='Temp in the grove!')
fig.show()
```

```python
fig = px.scatter(df, x="timestamp", y="humidity", title='Temp in the grove!')
fig.show()
```

```python
import matplotlib.pyplot as plt

# create a sample dataframe with temperature and humidity data
# create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))

# plot temperature data on the first subplot
ax1.scatter(df.timestamp, df['temperature'], color='red')
ax1.set_title('Temperature')

# plot humidity data on the second subplot
ax2.scatter(df.timestamp, df['humidity'], color='blue')
ax2.set_title('Humidity')

# set x-axis label on the second subplot
ax2.set_xlabel('Time')

# show the plot
plt.show()
```
